```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

```python
w_raw = "../weatherpy/weather_df.csv"
weather = pd.read_csv(w_raw)
xna_weather.head(20)
```

```python
# Configure gmaps
gmaps.configure(api_key = g_key)
# Locations
locations = xna_weather[["lat", "lng"]]
humidity = xna_weather["humidity"]
# Heatmap
fig = gmaps.figure()
# layer
heat = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=2)
# Add
fig.add_layer(heat)
# Display
fig
```

```python
no_clouds = xna_weather[xna_weather["cloudiness"] == 0]
above_70_degrees = no_clouds[no_clouds["maxtemp"] > 70]
below_80_degrees = above_70_degrees[above_70_degrees["maxtemp"] < 80]
slow_wind = below_80_degrees[below_80_degrees["windspeed"] < 10]
ideal_vacation = slow_wind[slow_wind["humidity"] < 60]
ind_vacation = ideal_vacation.reset_index()
del ind_vacation["index"]
ind_vacation
```

```python
gmaps.configure(api_key = g_key)
# Locations
locations = ind_vacation[["lat", "lng"]]
humidity = ind_vacation["humidity"]
# Heatmap
fig = gmaps.figure()
# layer
heat = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=2)
# Add
fig.add_layer(heat)

# Display
fig
```
